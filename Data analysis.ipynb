## Addressing the Research Question
### How many robberies were there and what were the outcomes?

```python
robberies = df1[df1.Crime.str.contains("Robbery")]
number_of_robberies = robberies.shape[0]
number_of_robberies
```

```python
robbery_outcomes = robberies.Outcome.unique()
robbery_outcomes
```

```python
for ro in robbery_outcomes:
    occurences = robberies[robberies.Outcome == ro]
    print('%4d %s' % (occurences.shape[0], ro))
```

### Robbery outcomes as a dataframe
This will enable the results to be sorted
Here we are making a different dataframe for outcome of robbery and its Occurances.

```python
cols_df2 = ['Outcome', 'Occurences']
df2 = pd.DataFrame(columns=cols_df2)
df2
```

```python
for ro in robbery_outcomes:
    occurences = robberies[robberies.Outcome == ro]
    df2 = df2.append({'Outcome':ro, 'Occurences': occurences.shape[0]}, ignore_index=True)
df2
```

```python
df3 = df2.sort_values('Occurences')
df3
```

Here we are sorting the dataframe on the basis of number occurences in increasing order.

## Further Work
Which crime are you most likely to go to prison for?

```python
cols_df3 = ['Crime', 'Occurences', 'Prison', 'Ratio']
df3 = pd.DataFrame(columns=cols_df3, dtype='str', index=None)

for crime in crimes:
    crime_count = df1[df1.Crime == crime].shape[0]
    crime_prison_count = df1[(df1.Crime == crime) & (df1.Outcome == 'Offender sent to prison')].shape[0]
    ratio = crime_prison_count / crime_count
    df3 = df3.append({'Crime':crime, 'Occurences':crime_count, 'Prison':crime_prison_count, 'Ratio': ratio}, ignore_index=True)
df3
```

```python
df4 = df3.sort_values('Prison', ascending=False)
plt.xticks(rotation=90)
plt.bar(df4.Crime, df4.Prison)
plt.show()
```

```python
df5 = df3.sort_values('Ratio', ascending=False)
plt.xticks(rotation=90)
plt.bar(df5.Crime, df5.Ratio)
plt.show()
```

From the Graph we can see that in case of shoplifting the chance of criminal being sent to prison is more than in other crimes. Theft from other person is less likely to be sent to prison
