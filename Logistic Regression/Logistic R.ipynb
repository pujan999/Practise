```python
# Logistic Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 447)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

numpy: NumPy is a general-purpose array-processing package. It provides a high-performance multidimensional array object, and tools for working with these arrays.</br>

matplotlib: Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.</br>

pyplot: matplotlib. pyplot is a collection of command style functions that make matplotlib work like MATLAB.</br>

    
    pd.read_csv: It retrieves data from csv file
dataset.iloc[:, :[2,3]].values:  It retrieves the datas from first to second and third column in the csv file. here the retrived value is assigned to X variable.</br>

dataset.iloc[:, 4].values: It retrieves the datas from forth columnin the csv file. here the retrived value is assigned to Y variable. </br>

sklearn.model_selection import train_test_split: Split arrays into random train and test subsets Here the x and y datasets are splitted for training and testing in ratio 1:4 </br>

StandardScaler: it is used for feature scaling. It changes the value of the feature by using the formula (x-Xmean)/sd of x and same for y fit_transform: it retrieves the new value and transforms the old value with new one.

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

LogisticRegression : This class implements regularized logistic regression using the ‘liblinear’ library,</br>

classifier.fit(X_train, y_train): This fits the data in X_train and Y_train in classification model</br>

 predict(): We can predict the class for new data instances using our finalized classification model in scikit-learn using the predict() function.</br>
 
Confusion matrix: This function creates a 2*2 matrix which helps to evaluate the predicion of classification model.

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
